```python
numbers = []
for i in range(10):
    n = int(input(f"Enter number {i+1}: "))
    numbers.append(n)
   
numbers.sort()
print("Smallest:", numbers[0])
print("Largest:", numbers[-1])
print("Second Largest:", numbers[-2])
```

```python
string = input("Enter a string: ")

vowels = consonants = digits = special = 0

for char in string:
    if char in "aeiouAEIOU":
        vowels += 1
    elif char.isalpha():
        consonants += 1
    elif char.isdigit():
        digits += 1
    else:
        special += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special:", special)
```

```python
t = (8, 5, 6, 9 , 4, 8 , 10)

print("Duplicates and their frequency:")
for item in set(t):
    count = t.count(item)
    if count > 1:
        print(f"{item} → {count} times")
```

```python
sentence = input("Enter a sentence: ")
words = sentence.split()

frequency = {}
for word in words:
    word = word.lower()
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print("Word Frequency:")
for word, count in frequency.items():
    print(f"{word} → {count}")
```

```python
students = {
    "khushmeet": 75,
    "nishiva": 92,
    "hemshikha": 60,
    "monika": 88,
    "purvi": 45
}

average = sum(students.values()) / len(students)
print(f"Average marks: {average}")

print("Students above average:")
for name, marks in students.items():
    if marks > average:
        print(f"{name}: {marks}")
```

```python
students = {}

while True:
    print("\n1. Add Student")
    print("2. View Students")
    print("3. Delete Student")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Name: ")
        marks = int(input("Marks: "))
        students[name] = marks
        print("Student added!")

    elif choice == "2":
        if students:
            for name, marks in students.items():
                print(f"{name} : {marks}")
        else:
            print("No students!")

    elif choice == "3":
        name = input("Enter name to delete: ")
        if name in students:
            del students[name]
            print("Deleted!")
        else:
            print("Not found!")

    elif choice == "4":
        print("the end")
        break
```

```python
num = int(input("Enter a number: "))
temp = num
digits = len(str(num))
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** digits
    temp //= 10

if total == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

```python
products = {}

while True:
    print("\n1. Add Product")
    print("2. View Products")
    print("3. Update Price")
    print("4. Delete Product")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Product name: ")
        price = float(input("Price: "))
        products[name] = price
        print("Product added!")

    elif choice == "2":
        if products:
            for name, price in products.items():
                print(f"{name}: ₹{price}")
        else:
            print("No products!")

    elif choice == "3":
        name = input("Product to update: ")
        if name in products:
            products[name] = float(input("New price: "))
            print("Updated!")
        else:
            print("Product not found!")

    elif choice == "4":
        name = input("Product to delete: ")
        if name in products:
            del products[name]
            print("Deleted!")
        else:
            print("Product not found!")

    elif choice == "5":
        print("Goodbye!")
        break
```
